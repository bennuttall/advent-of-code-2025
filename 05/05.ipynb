```python
from pathlib import Path

h, t = Path("input").read_text().split("\n\n")
```

```python
head = h.splitlines()
ingredients = {int(i) for i in t.splitlines()}

fresh_ingredient_ranges = set()
for row in head:
    a, b = row.split("-")
    r = range(int(a), int(b) + 1)
    fresh_ingredient_ranges.add(r)
```

```python
def check_ingredient_freshness(ingredient: int):
    for r in fresh_ingredient_ranges:
        if ingredient in r:
            return True
    return False
```

```python
print("Part 1:")
print(sum([check_ingredient_freshness(i) for i in ingredients]))
```

```python
intervals = [(r.start, r.stop - 1) for r in fresh_ingredient_ranges]

intervals.sort()

merged = []
for start, end in intervals:
    if not merged or start > merged[-1][1]:
        merged.append([start, end])
    else:
        merged[-1][1] = max(merged[-1][1], end)
```

```python
print("Part 2:")
print(sum(end - start + 1 for start, end in merged))
```
