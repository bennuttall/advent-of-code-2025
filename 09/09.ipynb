```python
from pathlib import Path

data = Path("input").read_text().splitlines()
```

```python
red_tiles_list = [tuple([int(n) for n in row.split(",")]) for row in data]
red_tiles = set(red_tiles_list)
```

```python
from itertools import combinations

def get_all_areas():
    for (a, b) in combinations(red_tiles, 2):
        x1, y1 = a
        x2, y2 = b
        yield (x2 - x1 + 1) * (y2 - y1 + 1)
```

```python
print("Part 1:")
print(max(get_all_areas()))
```

```python
from itertools import pairwise

green_edges = set()
for (a, b) in pairwise(red_tiles_list + [red_tiles_list[0]]):
    x1, y1 = a
    x2, y2 = b
    if x1 == x2:
        green_edges |= {(x1, y) for y in range(min(y1, y2) + 1, max(y1, y2))}
    elif y1 == y2:
        green_edges |= {(x, y1) for x in range(min(x1, x2) + 1, max(x1, x2))}
```
