```python
from pathlib import Path

data = Path("input").read_text().strip().splitlines()
```

```python
s = 0
for d in data:
    d1 = max(d[:-1])
    i = d.index(d1)
    d2 = max(d[i+1:])
    s += int(d1 + d2)
```

```python
print("Part 1:")
print(s)
```

```python
def largest_number(d: str, keep: int = 12):
    to_remove = len(d) - keep
    stack = []

    for digit in d:
        while to_remove and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)

    return int("".join(stack[:keep]))
```

```python
s = 0
for d in data:
    s += largest_number(list(d))
```

```python
print("Part 2:")
print(s)
```
