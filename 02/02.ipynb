```python
from pathlib import Path

data = Path("input").read_text().strip().split(",")
```

```python
def check_range(s: str) -> int:
    total = 0
    a, b = s.split("-")
    rng = range(int(a), int(b) + 1)
    assert len(rng)
    for r in rng:
        sr = str(r)
        if len(str(r)) % 2 == 0:
            if sr[:len(sr)//2] == sr[len(sr)//2:]:
                total += r
    return total
```

```python
print("Part 1:")
print(sum(check_range(d) for d in data))
```

```python
from functools import reduce

def factors(n):
    return set(reduce(
        list.__add__,
        ([i, n//i] for i in range(1, int(n**0.5) + 1) if n % i == 0))) - {n}

def is_silly(s: str) -> bool:
    for a in factors(len(s)):
        p = s[:a]
        b = len(s) // a
        if p * b == s:
            return True

def check_range_2(s: str) -> int:
    total = 0
    a, b = s.split("-")
    rng = range(int(a), int(b) + 1)
    assert len(rng)
    for r in rng:
        if is_silly(str(r)):
            total += r
    return total
```

```python
print("Part 2:")
print(sum(check_range_2(d) for d in data))
```
