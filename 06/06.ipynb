```python
from pathlib import Path

data = Path("input").read_text().strip().splitlines()
```

```python
rows = [row.split() for row in data]
```

```python
def compute(col: int):
    operator = rows[-1][col]
    total = int(rows[0][col])
    for row in rows[1:-1]:
        n = int(row[col])
        total = {
            '+': total + n,
            '*': total * n,
        }[operator]
    return total
```

```python
print("Part 1:")
print(sum(compute(col) for col in range(len(rows[0]))))
```

```python
OPERATORS_ROW = data[-1]
COLS = max(len(row) for row in data)
ROWS = len(data) - 1

def get_columns():
    nums = []
    for c in range(COLS):
        if len(OPERATORS_ROW) > c and OPERATORS_ROW[c] in "+*":
            operator = OPERATORS_ROW[c]
        digits = "".join(data[r][c] for r in range(ROWS) if data[r][c]).strip()
        if digits:
            nums.append(int(digits))
        else:
            yield (nums, operator)
            nums = []
    yield (nums, operator)
```

```python
def do_the_math():
    for nums, operator in get_columns():
        total, *nums = nums
        for n in nums:
            total = {
                '+': total + n,
                '*': total * n,
            }[operator]
        yield total
```

```python
print("Part 2:")
print(sum(do_the_math()))
```
