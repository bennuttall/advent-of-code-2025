```python
from collections import deque
```

```python
from pathlib import Path

instructions = Path("input").read_text().splitlines()
```

```python
dial = deque(range(100))
dial.rotate(50)
total = 0

for instruction in instructions:
    dir = instruction[0]
    steps = int(instruction[1:])
    if dir == "R":
        steps = - steps
    dial.rotate(steps)
    if dial[0] == 0:
        total += 1
```

```python
print("Part 1:")
print(total)
```

```python
dial = deque(range(100))
dial.rotate(50)
total = 0

for instruction in instructions:
    dir = instruction[0]
    steps = int(instruction[1:])
    for i in range(steps):
        n = -1 if dir == "R" else 1
        dial.rotate(n)
        if dial[0] == 0:
            total += 1
```

```python
print("Part 2:")
print(total)
```
