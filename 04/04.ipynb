```python
from pathlib import Path

data = Path("input").read_text().strip().splitlines()
```

```python
paper = set()
for y, row in enumerate(data):
    for x, c in enumerate(row):
        if c == "@":
            paper.add((x, y))
```

```python
def count_neighbours(x: int, y: int):
    neighbours = {
        (x-1, y-1), (x, y-1), (x+1, y-1),
        (x-1, y),           (x+1, y),
        (x-1, y+1), (x, y+1),  (x+1, y+1),
    }
    return sum([pos in paper for pos in neighbours])
```

```python
s = 0
for x, y in paper:
    if count_neighbours(x, y) < 4:
        s += 1
```

```python
print("Part 1:")
print(s)
```

```python
start = len(paper)
while True:
    before = len(paper)
    for x, y in list(paper):
        if count_neighbours(x, y) < 4:
            paper.remove((x, y))
    after = len(paper)
    if after == before:
        break
```

```python
print("Part 2:")
print(start - after)
```
