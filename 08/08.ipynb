```python
from pathlib import Path

data = Path("input").read_text().strip().splitlines()
```

```python
jbs = [tuple([int(x) for x in line.split(",")]) for line in data]
```

```python
import math
import networkx as nx

G = nx.Graph()
G.add_nodes_from(jbs)

def closest_jbs(G: nx.Graph) -> list[tuple[int, int]]:
    min_distance = float("inf")
    closest_pair = (None, None)

    for i in range(len(jbs)):
        for j in range(i + 1, len(jbs)):
            jb1 = jbs[i]
            jb2 = jbs[j]
            if G.has_edge(jb1, jb2):
                continue
            distance = math.sqrt(
                (jb1[0] - jb2[0]) ** 2 + (jb1[1] - jb2[1]) ** 2 + (jb1[2] - jb2[2]) ** 2
            )
            if distance < min_distance:
                min_distance = distance
                closest_pair = (jb1, jb2)

    return closest_pair

closest_jbs(G)
```

```python
def connect_closest_jbs(n: int):
    for i in range(n):
        a, b = closest_jbs(G)
        if (a, b) == (None, None):
            return
        G.add_edge(a, b)

connect_closest_jbs(1000)
```

```python
def mul(lst: list[int]) -> int:
    result = 1
    for num in lst:
        result *= num
    return result

circuits = sorted([(len(c), c) for c in nx.connected_components(G)], reverse=True)
```

```python
print("Part 1:")
print(mul([c[0] for c in circuits[:3]]))
```

```python
G2 = nx.Graph()
G2.add_nodes_from(jbs)

def connect_closest_jbs_until_one_circuit() -> tuple[tuple[int, int, int], tuple[int, int, int]]:
    while True:
        a, b = closest_jbs(G2)
        G2.add_edge(a, b)
        circuits = sorted([(len(c), c) for c in nx.connected_components(G2)], reverse=True)
        if len(circuits) == 1 and circuits[0][0] > 1:
            return (a, b)

a, b = connect_closest_jbs_until_one_circuit()
```

```python
print("Part 2:")
print(a[0] * b[0])
```
